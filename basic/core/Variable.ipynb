```python
# 자료형별 변수 초기화
dtype = 'A'   # 문자, 'A' 문자를 메모리에 할당하고 별명(변수)을 dtype이라고 지정
print(dtype)
dtype = 'KOREA SEOUL' # 문자열, 'KOREA SEOUL' 문자열을 dtype 변수에 덮어씀(새롭게 할당한다)
print(dtype)
dtype = 2022  # 정수, 2022 정수를 dtype 변수에 덮어씀
print(dtype)
dtype = 1.5   # 실수, 1.5 실수를 dtype 변수에 덮어씀
print(dtype)
dtype = True  # 논리형, True 논리형을 dtype 변수에 덮어씀
print(dtype)
```

```python
# 자료형별 변수 초기화
dtype = 'A'   # 문자, 'A' 문자를 메모리에 할당하고 별명(변수)을 dtype이라고 지정
print(type(dtype))
dtype = 'KOREA SEOUL' # 문자열, 'KOREA SEOUL' 문자열을 dtype 변수에 덮어씀(새롭게 할당한다)
print(type(dtype))
dtype = 2022  # 정수, 2022 정수를 dtype 변수에 덮어씀
print(type(dtype))
dtype = 1.5   # 실수, 1.5 실수를 dtype 변수에 덮어씀
print(type(dtype))
dtype = True  # 논리형, True 논리형을 dtype 변수에 덮어씀
print(type(dtype))
```

```python
# 일수를 입력받아 시, 분, 초로 변환하여 출력하세요.
# 1일 24시간, 1시간은 60분, 1분은 60초
# 3일은 72시간, 4320 분, 259200 초임
day = int(input('일수를 입력하세요: '))
hour = day * 24
minute = hour * 60
second = minute * 60
print(day, '일은', hour, '시간,', minute, '분,', second, '초임')
print(str(day) + '일은 ' + str(hour) + ' 시간, ' + str(minute) + ' 분, ' + str(second) + ' 초임')
print('{0}일은 {1} 시간, {2} 분, {3} 초임'.format(day, hour, minute, second))
print('%d일은 %d 시간, %d 분, %d 초임' % (day, hour, minute, second))
print(f'{day}일은 {hour} 시간, {minute} 분, {second} 초임') # python 3.6 부터 제공
```
