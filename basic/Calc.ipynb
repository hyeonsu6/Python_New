```python
# 1/4 분기 매출액 합을 출력하는 프로그램 제작, int(), input()

# 1월 매출액:1000
# 2월 매출액:1000
# 3월 매출액:1000
# 1/4분기 매출액: 3000 만원

mon1 = int(input('1월 매출액:'))
mon2 = int(input('2월 매출액:'))
mon3 = int(input('3월 매출액:'))

tot = mon1 + mon2 + mon3
print(f'1/4분기 매출액: {tot} 원')
```

```python
# m²/ 3.3 = 평

# 원룸의 width를 입력하세요(m):6
# 원룸의 height를 입력하세요(m):4
# 원룸의 크기: 24 m² 7.272727272727273 평

width = int(input('원룸의 width를 입력하세요(m):'))
height = int(input('원룸의 height를 입력하세요(m):'))
m2 = width * height
py = m2 / 3.3
print(f'원룸의 크기: {m2} m² {py} 평')
```

```python
# 신체질량지수(BMI) = (몸무게(kg) / 신장의 제곱) * 10000

# 몸무게(kg):64
# 키(신장, cm):167
# BMI: 22.948115744558788

kg = int(input("몸무게(kg):"))
cm = int(input("키(신장, cm):"))
bmi = (kg / cm**2) * 10000
print(f'BMI: {bmi}')
```

```python
# 지폐수 처리(변수, 연산자만 사용)
# 금액: 79460

# 50000원권: 1
# 10000원권: 2
# 5000원권: 1
# 1000원권: 4
# 500원권: 0
# 100원권: 4
# 50원권: 1
# 10원권: 1

wonkum = 79460
m50000 = wonkum // 50000
print(f'50000원권: {m50000}')

wonkum = wonkum - (50000 * m50000)
m10000 = wonkum // 10000
print(f'10000원권: {m10000}')

wonkum = wonkum - (10000 * m10000)
m5000 = wonkum // 5000
print(f'5000원권: {m5000}')

wonkum = wonkum - (5000 * m5000)
m1000 = wonkum // 1000
print(f'1000원권: {m1000}')

wonkum = wonkum - (1000 * m1000)
m500 = wonkum // 500
print(f'500원권: {m500}')

wonkum = wonkum - (500 * m500)
m100 = wonkum // 100
print(f'100원권: {m100}')

wonkum = wonkum - (100 * m100)
m50 = wonkum // 50
print(f'50원권: {m50}')

wonkum = wonkum - (10 * m50)
m10 = wonkum // 10
print(f'10원권: {m10}')
```

```python
# 토스뱅크의 보통예금은 년 2% 이자를 복리로 계산해준다. 금액을 입력했을 때 매일 받는 이자는?
# - 이자소득의 원천징수세율은 14%를 적용 할 것.

# 금액 입력: 50000000

# 원금: 50,000,000 원
# 년이자: 1,000,000 원
# 월이자: 83,333 원
# 오늘이자: 2,739 원
# 세후이자: 2,355 원

wonkum = 50000000

ija = int(wonkum * 0.02)
mon = int(ija / 12)
day = int(ija / 365)
receive = int(day - day * 0.14)

print(f'원금: {format(wonkum, ",")} 원')
print(f'년이자: {format(ija, ",")} 원')
print(f'월이자: {format(mon, ",")} 원')
print(f'오늘이자: {format(day, ",")} 원')
print(f'세후이자: {format(receive, ",")} 원')
```
