```python
# 고객이 1억 5천만원을 입금하면 1억까지는 2%를 지급하고 나머지 금액 5천만은 1%를 지급합니다.

# 금액(만원단위): 15000
# 원금이 1억을 초과합니다.
# ------------------------------
# 100,000,000 원 -> 2,000,000 원
# 50,000,000 원 -> 500,000 원

# 또는

# 금액(만원단위): 5000
# 원금이 1억 미만입니다.
# ------------------------------
# 50,000,000 원 -> 500,000 원

wonkum = int(input('금액(만원단위): '))
if wonkum > 10000:
    print('원금이 1억을 초과합니다.')
    print('-' * 30)
    won1 = 10000           # 1억
    won2 = wonkum - 10000  # 1억을 제외한 나머지 금액

    won1 = won1 * 10000;   # 만원단위 해제
    won2 = won2 * 10000;   # 만원단위 해제

    print(format(won1, ','), '원 ->', format(int(won1*0.02), ','), '원')
    print(format(won2, ','), '원 ->', format(int(won2*0.01), ','), '원')
else:
    print('원금이 1억 이하입니다.')
    print('-' * 30)
    won1 = wonkum

    won1 = won1 * 10000;
    print(format(won1, ','), '원 ->', format(int(won1*0.02), ','), '원')
```

```python
# 1. 성적을 입력받아 평균을 구하고 통과 여부를 판단하는 프로그램을 제작하시오.
# 60점 이상이면 [통과] 출력, 미만은 [다음에 응시해주세요.] 출력
# input 함수를 사용하지 말 것.

# Python 입사 점수: 95
# Java 입사 점수: 100
# Bigdata 입사 점수: 60
# 평균: 85.0
# 통과

python = 95
java = 100
bigdata = 60

tot = python + java + bigdata
avg = tot / 3
result=''

if avg >= 60:
  result = '합격'
else:
  result='다음에 응시해주세요.'

print(f'Python 입사 점수: {python}')
print(f'Java 입사 점수: {java}')
print(f'Bigdata 입사 점수: {bigdata}')
print(result)
print(f'평균: {avg}')
```

```python
# 1. 교보문고의 회원 등급을 확인하여 결재 금액을 출력하는 프로그램을 제작하시오.
# - 입력: 회원 등급, 도서 정가
# - 처리: 결재 금액 할인: P: 7% 할인, G: 6% 할인, S: 5% 할인, F: 2% 할인, 해당하는 등급이 없으면 'N'로 처리, 대소문자 구분됨
# - 출력: 회원 등급, 도서 정가, 결재 금액

# 회원 등급: P
# 도서 정가: 30,000
# 결재 금액: 27,900

# 회원 등급: N
# 도서 정가: 30,000
# 결재 금액: 30,000

grade = input('회원 등급: ')
price = int(input('도서 정가: '))
pay = 0

if grade == 'P':
  pay = price - (price * 0.07)
elif grade == 'G':
  pay = price - (price * 0.06)
elif grade == 'S':
  pay = price - (price * 0.05)
elif grade == 'F':
  pay = price - (price * 0.02)
else:
  grade = 'N'
  pay = price # 할인 없음

print('회원 등급:', grade)
print('도서 정가:', format(price, ','))
print('결재 금액:', format(int(pay), ','))
```

```python
# 2. 신체질량지수(BMI) = (몸무게(kg) / 신장의 제곱) * 10000 적용하여 계산후 등급을 산출하시오.
# - 입력:
#   몸무게(kg): 66
#   키(신장, cm): 167

# - 처리: 등급
#    . BMI: https://health.seoulmc.or.kr/healthCareInfo/myBMIPopup.do
#   BMI < 18.5: 저체중
#   BMI < 23: 정상
#   BMI < 25: 과체중
#   BMI >= 25: 비만

# 몸무게(kg): 66
# 키(신장, cm): 167.5
# BMI: 23.52 <- 소수 둘째 자리까지 출력, 반올림 처리
# 판정: 과체중

kg = int(input("몸무게(kg):"))
cm = float(input("키(신장, cm):"))
bmi = kg / (cm ** 2) * 10000

grade = ''

if bmi < 18.5:
    grade = '저체중'
elif bmi < 23:
    grade = '정상'
elif bmi < 25:
    grade = '과체중'
else:
    grade = '비만'

print(f'BMI: {bmi:.2f}')
print(f'판정: {grade}')
```

```python
# 전기 요금 계산기
# 사용량 kwh   200이하    201~400    400초과
# 단가(원)     99.3       187.9      280.6
# 기본요금(원) 910        1600       7300

# 전기 사용량을 입력받아 전기 요금을 계산하는 프로그램 제작
# 금액 = (사용량 * 단가) + 기본요금

# 전기 사용량을 입력: 300
# 사용량(kwh): 300
# 단가(원): 187.9
# 기본요금(원): 1,600
# 금액(원): 57,970

use = int(input('전기 사용량을 입력:'))

price=0 # 단가(원)
basic=0 # 기본요금(원)
if use <= 200:
    price = 99.3
    basic = 910
elif use >= 201 and use <= 400:
    price = 187.9
    basic = 1600
else: # elif use > 400:
    price = 280.6
    basic = 7300

tot = int((use * price) + basic)

print('사용량(kwh):', use)
print('단가(원):', price)
print('기본요금(원):', format(basic, ','))
print('금액(원):', format(tot, ','))
```
